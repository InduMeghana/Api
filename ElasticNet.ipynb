```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import ElasticNet
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
diabetes = load_diabetes()
diabetes_df = pd.DataFrame(diabetes.data , columns=diabetes.feature_names)
x = diabetes.data
y = diabetes.target
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 ,random_state = 45)
linear_model = LinearRegression(normalize = True)
linear_model.fit(x_train , y_train)
lasso_model = Lasso(alpha = 0.7 , normalize=True)
lasso_model.fit(x_train,y_train)
ridge_model = Ridge(alpha = 0.7 , normalize = True)
ridge_model.fit(x_train , y_train)
elastic_model = ElasticNet(alpha = 0.7 , l1_ratio = 0.5 ,normalize = True)
elastic_model.fit(x_train , y_train)
plt.figure(figsize=(10,8))
plt.plot(list(linear_model.coef_) + [linear_model.intercept_], 's--', alpha=0.7, color='red', label='Linear Model Weights')
plt.plot(list(ridge_model.coef_) + [ridge_model.intercept_], '*--', alpha=0.7, color='blue', label='Ridge Model Weights')
plt.plot(list(lasso_model.coef_) + [lasso_model.intercept_], '.--', alpha=0.7, color='magenta', label='Lasso Model Weights')
plt.plot(list(elastic_model.coef_)+[elastic_model.intercept_],'+--',alpha=0.7,color='black',label='ElasticNet Model Weights')
plt.grid()
plt.xlabel('Weights')
plt.ylabel('Values')
plt.legend()
plt.show()
plt.title("comparing R2-scores of models")
plt.pie([linear_model.score(x_test, y_test),ridge_model.score(x_test, y_test),lasso_model.score(x_test, y_test),elastic_model.score(x_test,y_test)],labels=["linear_model","ridge_model","lasso_model","elastic_net model"],colors=["white","blue","black","red"])
plt.show()
plt.title("comparing MSE of models")
plt.pie([mean_squared_error(y_train, linear_model.predict(x_train)),mean_squared_error(y_train, ridge_model.predict(x_train)),mean_squared_error(y_train, lasso_model.predict(x_train)),mean_squared_error(y_train,lasso_model.predict(x_train))],labels=["linear_model","ridge_model","lasso_model","elastic_net model"],colors=["white","blue","black","red"])
plt.show()
```
